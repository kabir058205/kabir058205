```python
# function without   return
def greet(firstname, lastname=None):
  if lastname:
    print(f"Hello {firstname} {lastname}")
  else:
    print(f"Hello {firstname}")

greet("Kabir")
greet("Kabir", "Singh")
```

```python
# function with return
def get_greeting(name):
  return f"Hello {name}"
message = get_greeting("Kabir")
print(message)
 
```

```python
def increment(number,by):
  return number + by
print(increment(2,1))
print(increment(2,by=1))
```

```python
def increment(number,by=1):
  return number + by
print(increment(2))
```

```python
def muliply(*numbers):
  total = 1
  for number in numbers:
    total *= number
  return total
print(muliply(2,3,4,5))
```
